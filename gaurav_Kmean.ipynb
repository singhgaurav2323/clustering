```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#importing dataset

data=pd.read_csv('Mall_Customers.csv')
X=data.iloc[:,3:5].values
```

```python
#getting value of k for k means
wcss=[]
X_range=[]
for i in range(1,14):
    clus=KMeans(n_clusters=i , init='k-means++',random_state=49)
    clus.fit(X)
    X_range.append(i)
    wcss.append(clus.inertia_)

#wcss curve
plt.plot(X_range,wcss)
plt.title('Elbow method')
plt.xlabel('number of cluster')
plt.ylabel('Wcss')
plt.show()
```

```python
#fitting to Kmean algorithum
kmeans=KMeans(n_clusters=5 , init='k-means++',random_state=49)
pred=kmeans.fit_predict(X)
```

```python
#plotting the result

plt.scatter(X[pred==0,0],X[pred==0,1],s=30,c='black',label='cluster 1')
plt.scatter(X[pred==1,0],X[pred==1,1],s=30,c='green',label='cluster 2')
plt.scatter(X[pred==2,0],X[pred==2,1],s=30,c='orange',label='cluster 3')
plt.scatter(X[pred==3,0],X[pred==3,1],s=30,c='blue',label='cluster 4')
plt.scatter(X[pred==4,0],X[pred==4,1],s=30,c='yellow',label='cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='orange',label='centroid',marker='*')
plt.legend()
plt.xlabel('Annual income k$')
plt.ylabel('rank on shopping')
plt.title('K-Mean Clustering')
plt.show()
```

```python
#plotting the result
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='orange',label='centroid',marker='*')
plt.legend()
plt.xlabel('Annual income k$')
plt.ylabel('rank on shopping')
plt.title('K-Mean Clustering')
plt.show()
```
